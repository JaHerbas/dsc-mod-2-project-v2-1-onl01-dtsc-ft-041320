```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('../model_df.csv')
df.head()
```

```python
sns.set()
sns.lmplot(x= 'time_since_renovated', y = 'price_o', data = df, line_kws={'color': 'red'})
```

While not a strong indicator, it does appear that recency of renovation does seem to effect sell price. We can see the the longer the house goes without renovation the price does make a gradual decline.
